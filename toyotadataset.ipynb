```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
final_car=car.drop('Model',axis=1)
```

```python
final_car.head()
```

```python
final_car['Current_year']=2020
```

```python
final_car.head()
```

```python
final_car['NoOf Years']=final_car['Current_year']-final_car['Mfg_Year']
```

```python
final_car.head()
```

```python
final_car.drop(['Mfg_Year','Current_year'],axis=1,inplace=True)
```

```python
final_car.head()
```

```python
Fuel= pd.get_dummies(final_car['Fuel_Type'],drop_first=True)
color = pd.get_dummies(final_car['Color'],drop_first=True)
```

```python
final_car =pd.concat([final_car,Fuel,color],axis=1)
```

```python
final_car.head()
```

```python
final_car.drop(['Fuel_Type','Color'],axis=1,inplace=True)
```

```python
final_car.head()
```

```python
corr=final_car.corr()
```

```python
final_car['Cylinders'].unique()
```

```python
X=final_car.drop('Price',axis=1)
y=final_car['Price']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
feat_imp = pd.Series(model.feature_importances_,index=X.columns)
feat_imp.nlargest(5).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random = RandomForestRegressor()
```

```python
n_estimators =[int(x) for x in np.linspace(start=100,stop=1200,num=12)]
max_features=['auto','sqrt']
max_depth= [int(x) for x in np.linspace(5,30,num=6)]
min_samples_split=[2,5,10,15,100]
min_samples_leaf=[1,2,5,10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_grid={'n_estimators':n_estimators,
             'max_depth':max_depth,
             'min_samples_split':min_samples_split,
             'min_samples_leaf':min_samples_leaf,'max_features':max_features}
print(random_grid)
```

```python
rf = RandomForestRegressor()
```

```python
rf_random=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,
                             scoring='neg_mean_squared_error',n_iter=10,cv=5,verbose=2,random_state=42,n_jobs=1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
pred = rf_random.predict(X_test)
```

```python
pred
```

```python
sns.distplot(y_test-pred)
```

```python
plt.scatter(y_test,pred)
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,pred)
```

```python
metrics.mean_squared_error(y_test,pred)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,pred))
```
